```
# --- TESTE INTEGRADO DO SISTEMA ---

# 1. Cadastro de Itens (Hash Table)
catalogo = CatalogoProdutos()
catalogo.inserir(ItemCardapio(10, 'X-Burger', 22.50))
catalogo.inserir(ItemCardapio(25, 'Batata G', 15.00))
catalogo.inserir(ItemCardapio(5, 'Suco Nat', 8.50))
catalogo.inserir(ItemCardapio(42, 'Combo Luxo', 55.00))

# 2. Uso do Carrinho (Stack)
meu_carrinho = CarrinhoCompras()
meu_carrinho.adicionar_item(catalogo.buscar(10))
meu_carrinho.adicionar_item(catalogo.buscar(42))

print(f"Carrinho atual: {[item.nome for item in meu_carrinho.itens]}")

# Simulação de erro do usuário -> Desfazer
removido = meu_carrinho.desfazer_ultimo()
print(f"[UNDO] Removido: {removido.nome if removido else 'Vazio'}")

# 3. Envio para Cozinha (Queue)
cozinha = ControleCozinha()
cozinha.novo_pedido(id_pedido=1, carrinho=meu_carrinho)

print(f"Fila da Cozinha: {len(cozinha.fila_pedidos)} pedido(s)")
pedido_preparo = cozinha.processar_proximo()
print(f"Preparando agora o pedido #{pedido_preparo['id']} com: {[i.nome for i in pedido_preparo['itens']]}")

# 4. Painel Administrativo (BST)
painel = PainelGerencial()
# Carregar todos os itens do catálogo no painel para análise de preços
for id_teste in [10, 25, 5, 42]:
    painel.inserir_produto(catalogo.buscar(id_teste))

print("\n--- RELATÓRIO DE PREÇOS (ORDENADO) ---")
for prod in painel.gerar_relatorio_ordenado():
    print(f"R$ {prod.preco:5.2f} | {prod.nome}")
```
